# Heat map

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tipscorr = tips[['total_bill','tip','size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(),center=True,annot=True)
```

## Cluster Map

```python
sns.clustermap(tipscorr.corr(),center=True,annot=True)
```
