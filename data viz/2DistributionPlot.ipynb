```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

We use distribution plots to visualize the distribution of quantitative data.
quantitative data is like information that can be expressed as a numerical value, meaning it can be counted or measured

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

* distplot/histplot
* jointplot
* pairplot
* rugplot

```python
# seaborn  load_datset 
#https://seaborn.pydata.org/generated/seaborn.load_dataset.html
df = sns.load_dataset('tips')
df
```

```python
sns.histplot(df['total_bill'],bins=20, kde=True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
#kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

**pairplot**

```python
sns.pairplot(df,hue='size',palette='rainbow')
```
